## Q3. Plot Time Vs Pulse as scatter plot and write your interpretation from the graph and type of correlation that you see?

We want to:

Plot time (x-axis) vs pulse (y-axis)

See the trend / type of correlation

Comment on relationship

Step 1: Convert time to numeric for plotting

Your time column is categorical (1 min, 15 min, 30 min).
We need numbers: 1, 15, 30.

```python
# Map time categories to numeric minutes
time_mapping = {'1 min': 1, '15 min': 15, '30 min': 30}
df['time_num'] = df['time'].map(time_mapping)
```

Step 2: Scatter Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df['time_num'], df['pulse'], alpha=0.7, color='blue')
plt.xlabel('Time (minutes)')
plt.ylabel('Pulse')
plt.title('Q3: Scatter plot of Time vs Pulse')
plt.grid(True)
plt.show()
```

Step 3: Pearson Correlation

```python
corr = df['time_num'].corr(df['pulse'])
print(f"Pearson correlation between Time and Pulse: {corr:.4f}")
```

Step 4: Interpretation (Brutal, Straight)

Correlation value (r):

r > 0 â†’ positive correlation (pulse increases with time)

r < 0 â†’ negative correlation

|r| < 0.2 â†’ weak / negligible linear correlation

Typical expectation for this dataset:

r â‰ˆ 0.5 â€“ 0.6 â†’ moderate positive correlation


What it means

r â†’ Pearson correlation coefficient (numeric measure of linear relationship between two variables)

r ranges from -1 to +1

r value	Interpretation
1	Perfect positive correlation
0.5â€“0.6	Moderate positive correlation
0	No linear correlation
-0.5 to -0.6	Moderate negative correlation
-1	Perfect negative correlation
In your dataset (Time vs Pulse)

If r â‰ˆ 0.5 â€“ 0.6 â†’ as exercise time increases, pulse tends to increase moderately.

Itâ€™s a positive linear relationship, but not perfect â€” some variation exists due to individual differences.

## Q5. Make a box plot using 'pulse' feature and indentify mean? Also, comment on the number of outlier visible?

Step 1: Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,5))
sns.boxplot(y=df['pulse'])
plt.title("Boxplot of Pulse")
plt.ylabel("Pulse")
plt.show()
```

Step 2: Compute Mean

```python
mean_pulse = df['pulse'].mean()
print(f"Mean of Pulse: {mean_pulse:.2f}")
```

Step 3: Identify Outliers (1.5Ã—IQR rule)

```python
# IQR calculation
Q1 = df['pulse'].quantile(0.25)
Q3 = df['pulse'].quantile(0.75)
IQR = Q3 - Q1

# Outlier fences
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR

# Extract outliers
outliers = df[(df['pulse'] < lower_fence) | (df['pulse'] > upper_fence)]
print(f"Number of outliers: {len(outliers)}")
print("Outlier values:")
print(outliers['pulse'].values)
```

âœ… Interpretation

Mean Pulse: Gives the central tendency (~99â€“100 bpm for this dataset).

Outliers:

Values below lower_fence or above upper_fence

Number of extreme pulse values is typically small

Distribution Shape:

If the median (line in box) â‰ˆ mean â†’ roughly symmetric

If mean > median â†’ right skew (some high pulse outliers)

ðŸ’¡ Brutal Truth:

Boxplot is the easiest way to spot unusual values in pulse

Outliers often correspond to measurement errors or unusual conditions

Most pulse values are in 80â€“120 bpm, outliers are rare

# Question 8: Make pairwise scatter of all features like following:


Now, use dimention reduction (e.g., PCA) and scatter plot the new features pairs. Now you will observe that only two features are sufficient to classify the iris data.

Step 1: Pairwise Scatter Plots (All Features)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot colored by species
sns.pairplot(iris, hue='species', diag_kind='kde', height=2.5)
plt.suptitle("Pairwise Scatter of Iris Features", y=1.02)
plt.show()
```

Explanation:

hue='species' â†’ colors by Iris species

diag_kind='kde' â†’ shows density distribution on the diagonal

This shows relationships between every pair of features, and you can visually see clusters forming.

Observation:

petal_length vs petal_width â†’ perfect separation

sepal features â†’ some overlap, harder to separate

Two features are almost enough to distinguish species.

```python
Step 2: PCA (Dimensionality Reduction)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 1. Extract features & scale
X = iris[features]  # sepal_length, sepal_width, petal_length, petal_width
X_scaled = StandardScaler().fit_transform(X)

# 2. Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 3. Create DataFrame for plotting
iris_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
iris_pca['species'] = iris['species']

# 4. Scatter plot of PCA components
plt.figure(figsize=(6,5))
sns.scatterplot(data=iris_pca, x='PC1', y='PC2', hue='species', s=80)
plt.title("PCA Scatter Plot (2 Components)")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.grid(True)
plt.show()
```

Step 3: Interpretation

PC1 and PC2 capture most of the variance (~95â€“96% combined).

Species clusters are well separated:

Setosa â†’ perfectly separated

Versicolor and Virginica â†’ small overlap in PC space

Insight:

Only two PCA features are sufficient to classify Iris species

Great for dimensionality reduction, ML models, and visualization

ðŸ’¡ Brutal Truth:

Raw 4D features â†’ hard to visualize

PCA â†’ compresses data while keeping almost all info

Makes classification & plotting extremely easy
